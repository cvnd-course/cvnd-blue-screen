```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
dog_image = cv2.imread('images/dog.jpg')

print("this is a image", type(dog_image),
     "with shape",dog_image.shape)
```

```python
%matplotlib inline

plt.imshow(dog_image)
```

```python
img_cpy = np.copy(dog_image)

img_cpy = cv2.cvtColor(img_cpy,cv2.COLOR_BGR2RGB)

plt.imshow(img_cpy)
```

```python
lower_blu = np.array([93,133,210])
upper_blu = np.array([216,216,255])

mask = cv2.inRange(img_cpy,lower_blu,upper_blu)

plt.imshow(mask,cmap='gray')
```

```python
masked_img = np.copy(img_cpy)
masked_img[mask!=0] = [0,0,0]
plt.imshow(masked_img)
```

```python
background = cv2.imread('images/fuji.jpg')
background_copy = np.copy(background)[0:480,0:852]
print(background_copy.shape,mask.shape)
background_copy = cv2.cvtColor(background_copy,cv2.COLOR_BGR2RGB)

plt.imshow(background_copy)
```

```python
background_copy[mask==0] = [0,0,0]
plt.imshow(background_copy)
```

```python
final_img = background_copy+masked_img
plt.imshow(final_img)
```
